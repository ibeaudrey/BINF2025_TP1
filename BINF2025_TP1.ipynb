*   Un scatterplot, ou nuage de points, montre comment une variable  est affectée par une autre. Les scatterplots sont souvent utilisés pour représenter une corrélation entre variables quantitatives qui semblent être liées et en deduire si leur relation est linéaire ou non-linéaire, forte ou faible, positive ou négative.
*   Un barplot, ou graphique à barres, est une représentation graphique qui utilise des barres rectangulaires pour comparer différentes catégories. Ils permettent aux utilisateurs de saisir rapidement les différences de taille ou de quantité entre les catégories.
*   Un boxplot, ou boîte à moustaches, est un moyen rapide de figurer le profil d'une série statistique quantitative. Ce diagramme est utilisé principalement pour comparer un même caractère dans deux populations de tailles différentes.
*   Un violinplot, ou diagramme en violon, est un graphique statistique permettant de comparer des distributions de probabilité.
*   Un histogramme est une représentation graphique permettant de représenter la répartition empirique d'une variable aléatoire ou d'une série statistique en la représentant avec des colonnes correspondant chacune à une classe et dont l'aire est proportionnelle à l'effectif de la classe1.
